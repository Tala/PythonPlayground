```python
def getColorSource():
    response = requests.get('https://en.wikipedia.org/wiki/List_of_colors_(compact)')
    html_doc = response.text
    soup = BeautifulSoup(html_doc, 'lxml')

    colors = list()
    main = soup.find('div', id='mw-content-text')
    for link in main.find_all('a'):
        colors.append(str(link.contents).replace('\'','').replace('[','').replace(']',''))
    return colors
```
