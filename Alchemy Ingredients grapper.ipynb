```python
from bs4 import BeautifulSoup
import urllib.request
with urllib.request.urlopen('http://www.worldofazolin.com/wiki/index.php?title=Herbalism_Recipe_List') as response:
   html_doc = response.read()

soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
links = list()

for link in soup.find_all('a'):
    links.append(str(link.get('href')))
    
links = [x for x in links if 'wiki' in x]
links = [x for x in links if not 'jpg' in x]
links
```

```python
titles = list()    
for link in links:
    with urllib.request.urlopen('http://www.worldofazolin.com' + link) as response:
        temp_doc = response.read()
        sub_soup = BeautifulSoup(temp_doc, 'html.parser')
        titles.append(sub_soup.title)
    
```
