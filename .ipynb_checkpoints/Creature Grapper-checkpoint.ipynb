```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

response = requests.get('https://en.wikipedia.org/wiki/Lists_of_legendary_creatures')
html_doc = response.text

soup = BeautifulSoup(html_doc, 'lxml')
```

```python
baseUrl = 'https://en.wikipedia.org'

links = list()
for link in soup.find_all('a'):
    links.append(str(link.get('href')))
    
links = [x for x in links if '/wiki/List_of_legendary_creatures_(' in x]
#links
```

```python
def makeTitlePretty(soupTitle):
    stringTitle = str(soupTitle)
    stringTitle = stringTitle.split('>')[1].split('-')[0]
    return stringTitle
```

```python
def get_content(url):
    response = requests.get(url)
    temp_doc = response.text
    sub_soup = BeautifulSoup(temp_doc, 'lxml')
    c1_title = sub_soup.title

    sub_links = list()
    for link in sub_soup.find_all('li'):
        sub_links.append(link.text)

        tempData = []
    for text in sub_links:
        try:
            description = text.split(' - ')[1]
            region = text.split('(')[1].split(')')[0]
            name = text.split(' (')[0]
            tempData.append([name, region, description])
        except:
            pass

    creature_data = pd.DataFrame(tempData, columns=['Name', 'Region', 'Description']) 

    return creature_data
```

```python
ingredients = dict()
for link in links:
    try:
        c_title, c_content = get_content('http://www.worldofazolin.com' + link)
        ingredients[c_title] = c_content
        print(c_title + ' done.')
    except:
        pass
```

```python
ingredients_dataframe = pd.DataFrame.from_dict(ingredients, orient='index')
ingredients_dataframe.to_excel('ingredients.xlsx')
```
