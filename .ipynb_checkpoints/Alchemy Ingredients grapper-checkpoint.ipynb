```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

with urllib.request.urlopen('http://www.worldofazolin.com/wiki/index.php?title=Herbalism_Recipe_List') as response:
   html_doc = response.read()

soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
links = list()
content = soup.find_all("div", id="content")
for link in content[0].find_all('a'):
    links.append(str(link.get('href')))
    
links = [x for x in links if '/wiki/index.php?title' in x]
links = [x for x in links if not 'Image' in x]
links = [x for x in links if not ':' in x]
links = [x for x in links if not 'Herbalism_Recipe' in x]
#links
```

```python
'http://www.worldofazolin.com' + links[1]
```

```python
def makeTitlePretty(soupTitle):
    stringTitle = str(soupTitle)
    stringTitle = stringTitle.split('>')[1].split('-')[0]
    return stringTitle
```

```python
def get_content(url):
    response = urllib.request.urlopen(url)
    temp_doc = response.read()
    sub_soup = BeautifulSoup(temp_doc, 'html.parser')
    c1_title = str(makeTitlePretty(sub_soup.title))

    content = sub_soup.find_all('p')
    c1_content = str(content[1:3]).replace('<p>','').replace('</p>','').replace('[','').replace(']','').replace('\n','')
    return c1_title, c1_content
```

```python
ingredients = dict()
for link in links:
    try:
        c_title, c_content = get_content('http://www.worldofazolin.com' + link)
        ingredients[c_title] = c_content
        print(c_title + ' done.')
    except:
        pass
```

```python
ingredients_dataframe = pd.DataFrame.from_dict(ingredients, orient='index')
ingredients_dataframe.to_excel('ingredients.xlsx')
```
